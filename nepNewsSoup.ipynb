```python
def paragraphOnlineKhabar(link):
    a = requests.get(link)
    para = ''
    a.encoding = 'UTF-8'
    a = a.text
    b = BeautifulSoup(a, 'lxml')
    m = b.find_all('p')
    for x in range(0, 2):
        para = para + m[x].text
    return para
```
