```python
ekantipur = requests.get('https://ekantipur.com/').text
```

```python
ekantipurSoup = BeautifulSoup(ekantipur, 'lxml')
# ekantipurSoup
```

```python
ekantipurSoup.title.text
```

```python
main_news = ekantipurSoup.find_all('' )
for x in main_news:
    print(x.prettify() )
    print()
```

```python
main_news = ekantipurSoup.find_all('a')
for x in main_news:
    type = x.get('data-type')
    if type == 'title':
        print(type)
        print(x)
        print()
    
```
